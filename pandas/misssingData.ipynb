```python
import pandas as pd
```

```python
df = pd.read_csv("weather_data.csv",parse_dates=['day'])
```

```python
df
```

```python
 df.set_index('day',inplace=True)
```

```python
df
```

```python
new_df = df.fillna(0)
```

```python
new_df
```

```python
new_df = df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no event'
})
```

```python
new_df
```

```python
#ffill is a forward fill where data from the previous rows were taken to fill 
# the data in NAN 
new_df = df.fillna(method='ffill')
new_df
```

```python
#bfill is a backward fill where data from the next rows were taken to fill 
# the data in NAN 
new_df = df.fillna(method='bfill')
new_df
```

```python
#This will copy the data column wise if we add axis = columns other 
# by default it is row
new_df = df.fillna(method = "bfill",axis='columns')
new_df
```

```python
#ffill is a forward fill where data from the previous rows were taken to fill 
# the data in NAN 
# adding limit means it will copy to specific rows like 1
new_df = df.fillna(method='ffill',limit=1)
new_df
```

```python
#interpolate will fill NAN values only numercial in a linear way so that data looks
#which better

new_df = df.interpolate()
new_df

# In interpolate we can mention other methods also like cubic, quadratic etc 
```

```python
new_df = df.interpolate(method = 'time')
new_df
```

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how='all')
new_df
```

```python
#thresh is threshold where we mention that if a row as 1 na value we will keep it
new_df = df.dropna(thresh=1)
new_df
```

# Replace Method to handle missing value of the dataset

```python
import pandas as pd
import numpy as np
# I am constatly changing the data in csv
```

```python
df = pd.read_csv("weather_data_NY.csv")
```

```python
df
```

```python
new_df = df.replace(-99999,np.NAN)
```

```python
new_df
```

```python
new_df = df.replace([-99999,-88888],np.NAN)
```

```python
new_df
```

```python
# in replace we can aslo use dictionary to replace the columns
df
```

```python
new_df = df.replace({
    'temperature':[-99999,-88888],
    'windspeed': -99999,
    "event": '0'
},np.NaN)
```

```python
new_df
```

```python
df
```

```python
new_df = df.replace({
    -99999: np.NaN,
    '0': 'Sunny'
})
```

```python
new_df
```

```python
df
```

```python
new_df = df.replace({
    'temperature': '[A-Za-z]',
    'windspeed':'[A-Za-z]'
},'',regex=True)
```
