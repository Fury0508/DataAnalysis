```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company = ['GOOGL','AMZN','MSFT','FB']
revenue = [90,136,89,27]
```

```python
ypos = np.arange(len(company))
plt.bar(ypos,revenue)
```

```python
plt.xticks(ypos,company)
plt.xlabel("US stocks")
plt.ylabel("renevue")
plt.bar(ypos,revenue,label = 'revenue')
plt.legend()
```

```python
company = ['GOOGL','AMZN','MSFT','FB']
revenue = [90,136,89,27]
profit = [40,2,34,12]
```

```python



plt.xticks(ypos,company)
plt.xlabel("US stocks")
plt.ylabel("renevue")
plt.bar(ypos-0.2,revenue,width = 0.4,label = 'revenue')
plt.bar(ypos+0.2,profit,width=0.4,label= 'profit')
plt.legend()
```

# Histograms

In histograms, x axis contains a variable and y axis will be a frequency of that variable

We have a sample data of blood sugar level of different patients, we will try to plot number of patients by blood range and try to figure out how many patients are normal, pre-diabetic and diabetic

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, rwidth=0.8) # by default number of bins is set to 10
```

```python
plt.hist(blood_sugar, bins = 3,rwidth=0.8)
```

```python
plt.hist(blood_sugar, bins = [80,100,125,150],rwidth=0.8)
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95,
 color=['green','orange'],label=['men','women'])
plt.legend()
```

```python
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95,
 color=['green','orange'],label=['men','women'],orientation='horizontal')
```
